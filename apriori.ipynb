## Importing the libraries

```
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Data Preprocessing

```
dataset = pd.read_csv('Market_Basket_Optimisation.csv', header = None)  #header none eklememizin sebebi bizim datasetimize bakarsak öncekilerde her zaman columnların belli bir adı vardı fakat şu an ki datasetimizde her column aynı şeyleri ifade etmiyor yani içinde random şeyler var bu yüzden column name koymamamız lazım hicbi columna.
transactions = []
for i in range(0,7501):   #burda 7501 dememizin sebebi 7500 e kadar datalarımızı dolduracağız datasetimiz böyle.
  transactions.append([str(dataset.values[ i, j ])  for j in range(0,20)]) #list of product yapmak için boş köşeli parantez koyuyoruz.
  #str(dataset.values[ i, j ] bunu dememizin sebebi apriori datalarının içinde sadece string olmak zorunda biz de str ile string olmaya zorluyoruz. .values da kullanmamız gerekiyor çünkü appendle içine doldururken bunu kullanıyoruz.
  #i ve j dediğimizde ilki row ikincisi column oluyor datasetimizi çektik ve rowları for loopunun içindeki i ile çektiğimiz için birinciye i dedik 20 tane column olduğu için ve onu da ikinci for loopundaki j yi kullanarak oluşturduğumuz için oraya da j yazdık.
  #tek satır for loopunda içini dolduracağımız şeyi başa yazıyoruz.
```

## Training the Apriori model on the dataset

```
from apyori import apriori
rules = apriori(transactions = transactions , min_support = 0.003, min_confidence = 0.2, min_lift = 3, min_lenght = 2, max_lenght = 2)  #apriori içindeki ilk parametremiz datasetten oluşturduğumuz listemiz oluyor. ikinci parametre ise minimum support değerini belirlememiz için. belirlediğimiz bu değerin altındaki değerleri dikkate almayacağız.
#supportta min belirlerken bizim istediğimiz şey kuraldan hatırladıgımız uzere 2 ürünün ne kadar çok birlikte alındığı. biz 2 ürünün günde en az 3 kere birlikte alındığı senaryoyu istiyoruz. bu yüzden (3x7(datamız 1 haftalık oldugu icin 1 günü bulmak istiyoruz 7 yle de çarpıcaz))/all of transactions
#minconfidence için bizim default valuemuz 0.8
#minlift için önerilen de 3
#biz burda 1 alana 1 bedava yaptıgımız icin 2 ürün istiyoruz o yüzden min ve max lenght i 2 olarak belirledik.
```

## Visualising the results

### Displaying the first results coming directly from the output of the apriori function

```
results = list(rules)
results  #burda ilk itemleri alanların ikincileri alma şansı yüksek oluyor.
```

### Putting the results well organised into a Pandas DataFrame

```
def inspect(results):
    lhs         = [tuple(result[2][0][0])[0] for result in results]
    rhs         = [tuple(result[2][0][1])[0] for result in results]
    supports    = [result[1] for result in results]
    confidences = [result[2][0][2] for result in results]
    lifts       = [result[2][0][3] for result in results]
    return list(zip(lhs, rhs, supports, confidences, lifts))
resultsinDataFrame = pd.DataFrame(inspect(results), columns = ['Left Hand Side', 'Right Hand Side', 'Support', 'Confidence', 'Lift'])
```

### Displaying the results non sorted

```
resultsinDataFrame
```

### Displaying the results sorted by descending lifts

```
resultsinDataFrame.nlargest(n = 10, columns = 'Lift')
```
